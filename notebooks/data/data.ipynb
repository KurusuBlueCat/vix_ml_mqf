```python
T10Y2Y = pd.read_csv('T10Y2Y.csv')
T10Y2Y.DATE = T10Y2Y.DATE.apply(pd.to_datetime)
T10Y2Y.set_index('DATE',inplace=True)
T10Y2Y = T10Y2Y.loc['1999-11-19':,:]
T10Y2Y 
```

```python
drop_list = T10Y2Y[T10Y2Y.T10Y2Y == '.'].index.to_list()
T10Y2Y.drop(drop_list,axis=0,inplace=True)
T10Y2Y = T10Y2Y.astype('float')
T10Y2Y 
```

```python
gold = pdr.get_data_yahoo('GC=F',start=start,end=end)
gold
```

```python
HSI_1 = pd.read_csv('HSI_1.csv')
HSI_1 = HSI_1[::-1]
HSI_1
```

```python
HSI_2 = pd.read_csv('HSI_2.csv')
HSI_2 = HSI_2[::-1]
HSI_2
```

```python
HSI = pd.concat([HSI_1,HSI_2],axis=0)
HSI.Date = HSI.Date.apply(pd.to_datetime)
HSI.set_index('Date',inplace=True)
HSI
```

```python
inflation = pd.read_csv('inflation.csv')
inflation.TIME= inflation.TIME.apply(pd.to_datetime)
inflation.set_index('TIME',inplace=True)
```

```python
inf_data = pd.DataFrame(inflation,index=pd.date_range('11/1/1999','11/22/2022')).ffill()
inf_data = inf_data.loc['1999-11-19':,:]
inf_data = pd.DataFrame(inf_data,index=vix.index)
```
